#remove-html-tags

```python
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
text = '<sdafasd> </fasdfasdfa> asdfasdga <asdfgasdf>asdga sdg. asdgasdgasdgas <asd>asfgfasdasdgsdgdsgasd<gasdgas>aadgsgsgASDgadgasad'
```

```python
remove_html_tags(text)
```

```python
df.loc[:, 'review'] = df['review'].apply(remove_html_tags)
```

#remove-url

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
text = 'https://www.google.com/ asgasdfga fsg aeg adfg dfg dfsdfhsdfhsfdhsdfh https://www.google.com/'
```

```python
remove_url(text)
```
